# 02. Finding aggregated mean of "order_number" by "department_id"

```python
# calculating the mean 'order_number' grouped by 'department_id' of 'df' subset
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# calculating the mean 'order_number' grouped by 'department_id' of entire df
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

# 03. Observations

In both the subset and entire database, department id 21 (missing category) has the highest mean value meaning that quite a bit of the products in instacart have a missing category id and are therefore missing valuable consumer information.

In the entire database, department id 5 (alcohol cateogry) has the lowest average number of orders per user across all departments. This could be an interesting area to delve IF we had age information on users. However we do not, but this is not surpising as some users are probably less than 21 and certain retailers do not allow alcohol purchases on Sundays or altogether.

In the subset, department id 12 (meat / seafood category) has the lowest average number of orders per user across all departments. This is surprising to me and indicates the subset does not represent the overall dataframes insights.

# 04. Loyalty flag for existing customers

```python
# create "max_order" column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# inspecting new column
ords_prods_merge.head()
```

```python
# assigning loyalty flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# "loyalty flag" frequency check
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```

# 05. Customer spending habits

```python
# pulling basic statistics
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'order_number': ['mean']})
```

I can initially see that the max value is most likely incorrect across the board, so it is difficult to glean insight from the max metric. For the min metric, I am not surprised it is $1 across the board and doesn't tell me much. The only real information comes from the mean column, where we can see that Loyal customers spend less than both the new and regular customers. They also place much more orders compared to the new and regular customers, meaning they order more and spend less per product.

# 06. Different types of spenders

```python
# creating "average_spend" column
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# inspecting new column
ords_prods_merge.head()
```

```python
# assigning loyalty flags
ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# "spending flag" frequency check
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id','average_spend', 'spending_flag']].head(20)
```

# 07. Determining frequent and non-frequent customers

```python
# creating "customer_frequency" column
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# inspecting new column
ords_prods_merge.head()
```

```python
# assigning loyalty flags
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['customer_frequency'] <= 20) & (ords_prods_merge['customer_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# "frequency_flag" frequency check
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

The 5 NaN users should indicate there are five users who have only placed one order.
