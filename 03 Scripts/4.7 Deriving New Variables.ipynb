```python
# define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

### 02. Creating ' two busiest' and 'two slowest' days columns

```python
# define the loop function

result_2 = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 3 or value == 4:
    result_2.append("Least busy days")
  else:
    result_2.append("Regularly busy")
```

```python
# creating a new column within the df
ords_prods_merge['busiest_days'] = result_2
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

### 03. Accuracy and Observations

The two busiest days are Saturday and Sunday (the weekend) and the two slowest days are Tuesday and Wednesday. The sum of rows for the busiest and slowest days match up to their individual days so that tells me it is accurate.

### 04. Identify the busiest hours of the day

```python
ords_prods_merge.columns
```

By reviewing the column names I will use the 'order_hour_of_day' value to reference for the loop function

```python
# frequency check for 'order_hour_of_day'
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

By referencing the frequency table above I will divide the three categories with eight hours each based on ascending order.

Most orders: 10, 11, 14, 15, 13, 12, 16, 9

Average orders: 17, 8, 18, 19, 20, 7, 21, 22

Fewest orders: 23, 6, 0, 1, 5, 2, 4, 3

```python
# define the loop function

result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10,11,14,15,13,12,16,9]:
        result_3.append("Most orders")
    elif value in [17,8,18,19,20,7,21,22]:
        result_3.append("Average orders")
    else:
        result_3.append("Fewest orders")
```

```python
# creating a new column within the df
ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
